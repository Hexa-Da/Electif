<div class="alert alert-block alert-info">
    <p> Écrivez une fonction "make_dict" qui prend en paramètre le nom de deux colonnes d'un dataframe, et renvoie un dictionnaire des éléments de la première vers ceux de la seconde.</p>
    <p>Utilisez cette fonction pour créer un dictionnaire associant chacune des colonnes "Id Grande discipline", "Id Discipline" et "Id Secteur disciplinaire" respectivement aux colonnes "Grande discipline", "Discipline" et "Secteur disciplinaire". Affichez-les afin de les vérifier.</p>
    <p><strong>Aide</strong> : consultez la documentation des fonctions "drop_duplicates", "set_index" et "to_dict".</p>
</div>

```python
def make_dict(col1, col2):
    return df_eleves[[col1, col2]].drop_duplicates().set_index(col1).to_dict()[col2]

gd_disc_dict = make_dict('Id Grande discipline', 'Grande discipline')
print('\nGrande discipline : ')
print(gd_disc_dict)

disc_dict = make_dict('Id Discipline', 'Discipline')
print('\nDiscipline : ')
print(disc_dict)

sect_disc_dict = make_dict('Id Secteur disciplinaire', 'Secteur disciplinaire')
print('\nSecteur disciplinaire : ')
print(sect_disc_dict)
```

## Relation entre sexe et validation de L1 par filière.

On va maintenant regarder la relation entre le sexe des étudiant·e·s et la validation de la L1.

<div class="alert alert-info">
    Reproduisez l'image ci-dessous.
</div>



```python
# Statistiques pour la partie droite

grouped = df_eleves.groupby(['Id Grande discipline', 'Id Sexe'])\
    .sum()[["Effectif de néobacheliers de la cohorte", "Passage en L2 en 1 an", "Passage en L2 en 1 ou 2 ans"]]
statistiques_sexe = grouped
statistiques_sexe["ratio_1_an"] = grouped["Passage en L2 en 1 an"]/grouped["Effectif de néobacheliers de la cohorte"]
statistiques_sexe["ratio_2_an"] = grouped["Passage en L2 en 1 ou 2 ans"]/grouped["Effectif de néobacheliers de la cohorte"]
statistiques_sexe
```

```python
# Statistiques pour la partie gauche
data2 = df_eleves.groupby(['Id Grande discipline', 'Id Sexe'])[['Effectif de néobacheliers de la cohorte',
                                                               'Passage en L2 en 1 an',
                                                               'Passage en L2 en 1 ou 2 ans']].sum()
data2['Passage en L2 en 2 ans'] = data2['Passage en L2 en 1 ou 2 ans']- data2['Passage en L2 en 1 an']
data2['Echec après 2 ans'] = data2['Effectif de néobacheliers de la cohorte']- data2['Passage en L2 en 1 ou 2 ans']
data2['Passage en L2 en 1 an'] = data2['Passage en L2 en 1 an']/data2['Effectif de néobacheliers de la cohorte']*100
data2['Passage en L2 en 2 ans'] = data2['Passage en L2 en 2 ans']/data2['Effectif de néobacheliers de la cohorte']*100
data2['Echec après 2 ans'] = data2['Echec après 2 ans']/data2['Effectif de néobacheliers de la cohorte']*100
data2
```
